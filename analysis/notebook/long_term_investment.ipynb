```python
FEE = 5.0 * 2
POCKET_FEE = 2.0 * 2
LARGE_FEE = 10.0
```

```python
# Portfolio 1
TOTAL = 3000
PURCHASE_PRICE = {
    'NEC' : 1.93,
    'IFT' : 9.4,
}
is_pocket = {
    'NEC' : False,
    'IFT' : False
}
STOCK_CODES = ["NEC", "IFT"]
```

```python
result = ef_random_portfolio(STOCK_CODES, days = 365)
for stock_code in result:
    print(f"Code: {stock_code}")
    fee = FEE if not is_pocket[stock_code] else POCKET_FEE
    investment = TOTAL * result[stock_code]
    print(f"Investment: {investment}")
    if investment > fee:
        print(f"No of shares: {(investment - fee) / PURCHASE_PRICE[stock_code]}")
    else:
        print("Just don't invest :).")
```
